# EDA for Census Data

Simple EDA for exploring patterns and assess data quality issues in given [census income dataset](https://archive.ics.uci.edu/ml/datasets/census+income).


```python
# imports
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Get Census Data

```python
raw_df = pd.read_csv("../data/census.csv")
raw_df.head()
```

## General Inspection and Summary Statistics

```python
raw_df.info()
```

```python
raw_df.describe()
```

```python
target = " salary"
num_cols = raw_df.select_dtypes(include=[np.number]).columns
cat_cols = list(set(raw_df.columns) - set(num_cols))
```

```python
sns.histplot(raw_df[target])
plt.show()
```

No missing data. Imbalanced categorical distribution on target. Stratification on data split would be required as well as model should take imbalanced class into account.

## Relations with Target and Features

```python
for col in cat_cols:
    sns.countplot(data = raw_df, hue=target, x=col)
    plt.xticks(rotation=45)
    plt.show()
```

```python
for col in num_cols:
    sns.histplot(data = raw_df, hue=target, x=col)
    plt.xticks(rotation=45)
    plt.show()
```
